```python
import pandas as pd
import os

df = pd.read_csv("SafeSurf.csv")

def read_dataset(file_map):
    for i, col in enumerate(df.columns):
            print(f"{i+1}. {col}")
    
            

file = "SafeSurf.csv"

read_dataset(file)
```

```python
df.info()
print(df.isnull().sum())
```

```python
df = df.drop(['Src IP', 'Src Port', 'Dst IP', 'Dst Port', 'Timestamp'], axis=1)
```

```python
import numpy as np
df = df.replace([np.inf, -np.inf], np.nan).dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Label'])
print(f"Labels converted: {dict(zip(le.classes_, le.transform(le.classes_)))}")
```

```python
non_zero_cols = df.columns[(df != 0).any(axis=0)]
dropped_zero_cols = set(df.columns) - set(non_zero_cols)
df = df[non_zero_cols]
```

```python
# Sütun isimlerinde 'flag' kelimesi geçenleri (büyük/küçük harf fark etmeksizin) bul
flag_cols = [col for col in df.columns if 'flag' in col.lower()]

print(f"Silinen {len(flag_cols)} adet Flag sütunu şunlar:")
print(flag_cols)

# Bu sütunları veri setinden kalıcı olarak at
df = df.drop(columns=flag_cols)

# Kontrol: Yeni sütun sayısı
print(f"\nGüncel Sütun Sayısı: {len(df.columns)}")
```

```python
df = df.drop(['Down/Up Ratio', 'Fwd Pkts/b Avg', 'Fwd Blk Rate Avg', 'Bwd Pkts/b Avg', 'Bwd Blk Rate Avg', 
        'Fwd Seg Size Min'], axis=1)
```

```python
drop_list = [
    'Bwd Seg Size Avg',      
    'Subflow Fwd Pkts',      
    'Subflow Bwd Pkts',      
    'Subflow Fwd Byts',     
    'Subflow Bwd Byts',     
    'Fwd Byts/b Avg',        
    'Bwd Byts/b Avg'        
]

drop = [col for col in drop_list if col in df.columns]

df = df.drop(columns=drop)
```

```python
last_drop = ['Pkt Size Avg', 'Fwd IAT Tot', 'Bwd IAT Tot']
cols_to_drop = [c for c in last_drop if c in df.columns]
df = df.drop(columns=cols_to_drop)
```

```python
initial_rows = len(df)
df = df[ (df['Tot Fwd Pkts'] + df['Tot Bwd Pkts']) > 1 ]
```

```python
print(f"Tek paketlik gürültüler silindi. Silinen: {initial_rows - len(df)}")
```

```python
before_duration = len(df)
df = df[ df['Flow Duration'] > 0 ]
```

```python
print(f"0 süreli hatalı kayıtlar silindi. Silinen: {before_duration - len(df)}")
```

```python
before_dup = len(df)
df = df.drop_duplicates()
```

```python
print(f"Tekrar eden satırlar silindi. Silinen: {before_dup - len(df)}")
```

```python
print("\n--- Sınıf Dağılımı (Label) ---")
print(df['Label'].value_counts())
```

```python
df.to_csv("csharp_final_data.csv", index=False)
```
